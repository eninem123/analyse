```python
import pandas as pd
```

```python
unames = ['user_id','gender','age','occupation','zip']
```

```python
users = pd.read_table('./pydata-book-master/ch02/movielens/users.dat',sep='::',header=None,names=unames)
```

```python
users.head()
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
```

```python
ratings = pd.read_table('./pydata-book-master/ch02/movielens/ratings.dat', sep='::', header=None, names=rnames)
```

```python
ratings.head()
```

```python
mnames = ['movie_id', 'title', 'genres']
```

```python
movies = pd.read_table('./pydata-book-master/ch02/movielens/movies.dat',sep='::', header=None, names=mnames)
movies
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
# 合并三个表
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data.head()
```

```python
data.ix[0]
```

```python
# 按照性别计算每部电影的平均得分
mean_ratings = data.pivot_table('rating', index='title',columns='gender',aggfunc='mean')
```

```python
mean_ratings[:5]
```

```python
# 想过滤掉评分不达250条的电影
# 先对电影分组，有多少电影
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings
```

```python
# 为了解女性观众最喜欢的电影，对f列降序排列,by的作用是对某列排序
```

```python
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
```

```python
top_female_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
```

```python
sorted_by_diff[:15]
```

```python
# 对结果反序再取前15行,得到男性观众更喜欢的电影
sorted_by_diff[::-1][:15]
```

```python
# 如果想找出分歧最大的电影,可以根据方差标准差
ratings_std_by_title = data.groupby('title')['rating'].std()
ratings_std_by_title
```

```python
# 根据active_titles进行过滤
ratings_std_by_title = ratings_std_by_title.ix[active_titles]
ratings_std_by_title
```

```python
# 根据值进行降序，标注差越大，分歧越大
ratings_std_by_title.sort_index(ascending=False)[:10]
```
